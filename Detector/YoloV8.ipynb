The results of the object detection are saved to runs\detect\predict. Lets visualize some of these predictions.

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
path = "runs/detect/predict"
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# List of image file paths
image_paths = ['image1.jpg', 'image2.jpg', 'image3.jpg', 'image4.jpg', 'image5.jpg']

# Create a figure to hold the subplots
fig, axes = plt.subplots(1, len(image_paths), figsize=(15, 5))  # Adjust the figsize as needed

# Loop through images and display them side by side
for i, image_path in enumerate(image_paths):
    img = mpimg.imread(image_path)
    axes[i].imshow(img)
    axes[i].axis('off')  # Turn off axis labels for cleaner display

# Show the figure
plt.show()
```

```python
img_paths = list(os.listdir(path))
```

```python
import random
random.shuffle(img_paths)
```

```python
count = 0
fig, axes = plt.subplots(1, 5, figsize=(15, 5))  # Adjust the figsize as needed

for i in img_paths:
    if count == 5:
        break
    img_path = path + '/' + i
    img = Image.open(img_path)
    axes[count].imshow(img)
    axes[count].axis('off')  # Turn off axis labels for cleaner display
    count += 1
```

It seems we are having difficulty finding the bounding boxes for clear waste items, like plastic containers and bottles. Let's take a look at some of the metrics we have from the training.

```python
metric_path = 'runs/detect/train32'

confusion = Image.open(metric_path + '/' + 'confusion_matrix_normalized.png')
plt.figure(figsize=(10, 10))
plt.imshow(confusion)
plt.axis('off')
```

```python
results = Image.open(metric_path + '/' + 'results.png')
plt.figure(figsize=(10, 10))
plt.imshow(results)
plt.axis('off')
```

The YoloV8 model seems to have room for improvement. With more compute resources, this model could definitely be improved with more training epochs.
